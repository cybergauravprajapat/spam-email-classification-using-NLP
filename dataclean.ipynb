```python
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer, WordNetLemmatizer
from textblob import TextBlob
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
datset = pd.read_csv("email.csv")
```

```python
datset.head(2)
```

```python
datset['Message'] = datset['Message'].str.lower()
```

```python
number = string.digits
sp_char = string.punctuation
def remove_number_sp_char(data):
    new_text = ""
    for i in data :
        if (i not in number) and (i not in sp_char):
            new_text += i
    return new_text
datset['Message'] = datset['Message'].apply(remove_number_sp_char)
```

```python
stopwords_list = stopwords.words('english')
def stopwords_remove(data):
    new_text = ""
    for i in word_tokenize(data):
        if i not in stopwords_list :
            new_text = new_text+" "+i
    return new_text[1:]
datset['Message'] = datset['Message'].apply(stopwords_remove)
```

```python
def Lemmatizer_text(data):
    new_text =""
    for i in word_tokenize(data):
        lem = WordNetLemmatizer()
        text = lem.lemmatize(i,pos="a")
        new_text = new_text + " "+text
    return new_text[1:]
datset['Message'] = datset['Message'].apply(Lemmatizer_text)
```

```python
tdf1 = TfidfVectorizer(max_features=4000)
new_data = pd.DataFrame(tdf1.fit_transform(datset['Message']).toarray())
```
